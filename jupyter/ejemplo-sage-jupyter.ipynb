# Ejemplo pr√°ctico estudio funciones

Para este ejemplo haremos uso de 
$$f(x)=\frac{x+3}{x}.$$ 

```python
f= (x+3)/x
```

Para poder representarla, hacemos uso de `plot`.

```python
plot(f,(-10,10),ymax=10, ymin=-10)
```

La funci√≥n no est√° definida en 0. Por tanto el dominio de definici√≥n de $f$ es $\mathbb{R}\setminus\{0\}$.

En cuanto a los puntos de corte con el eje OX, lo podemos encontrar con `solve`.

```python
solve(f,x)
```

# Continuidad

La funci√≥n es cont√≠nua en todo su dominio, pues es un cociente de dos polinomios, y en el dominio el denominador no se anula. Podemos estudiar los l√≠mites laterales en el 0.

```python
f.limit(x=0,dir="-")
```

```python
f.limit(x=0,dir="+")
```

# As√≠ntotas

Por lo visto en el apartado anterior, tenemos que $f$ tiene una as√≠ntota vertical en 0. Veamos si tiene as√≠ntotas horizontales. 

```python
f.limit(x=oo, dir="+")
```

```python
f.limit(x=-oo, dir="-")
```

Por lo que $f$ tiene una as√≠ntota horizontal en $y=1$.

```python
plot(f,(-10,10),ymax=10, ymin=-10)+plot(1,(-10,10), linestyle="--", color="gray")
```

# Crecimiento

Para estudiar la monoton√≠a de la funci√≥n calculamos su derivada y d√≥nde se anula.

```python
df=diff(f,x)
expand(df)
```

```python
solve(df,x)
```

Al no anularse, basta ver su valor en un punto cualquiera para ver si es estrictamente positiva o negativa.

```python
df.subs({x:1})
```

De esta forma obtenemos que $f$ es estrictamente decreciente en todo su dominio. 

Como la derivada no se anula, no tiene ni m√°ximos ni m√≠nimos relativos. 

# Convexidad

Veamos en qu√© signo tiene la segunda derivada.

```python
ddf=diff(df,x)
expand(ddf)
```

Como vemos, para $x<0$, $f''(x)<0$, y para $x>0$, $f''(x)>0$. Por tanto, $f$ ser√° c√≥ncava en $(-\infty,0)$ y convexa en $(0,\infty)$.

# Paridad

La funci√≥n carece de paridad, pues $f(x)\neq -f(x)$ y $f(x)\neq f(-x)$.

```python
plot(f,(-10,10),ymax=10, ymin=-10)+\
plot(f.subs({x:-x}),(-10,10), linestyle="--", color="green")+\
plot(-f,(-10,10), linestyle="--", color="red")
```
